### import matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5,11)
y = x ** 2
```

```python
x
```

```python
y
```

Operating Matplotlib using Functions

```python
plt.plot(x,y)
```

To Label the axis we use:

```python
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Output VS Input')
```

Multiple Plots

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

OOP Matplotlib method

```python
fig = plt.figure()
# this takes in 4 paprameters
axes = fig.add_axes([0.1,0.1,0.8,0.8])

# to plot we use:
axes.plot(x,y)

# to label and add a title we use:
axes.set_xlabel('input')
axes.set_ylabel('output')
axes.set_title('title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.9])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')
axes2.plot(y,x)
axes2.set_title('SMALLER INVERSE PLOT')
```
